```
#Created a new smaller dataframe "df2"
df2 = df[["UNITID", "INSTNM", "CITY", "STABBR", "ZIP", "ADM_RATE", "UGDS", "TUITIONFEE_IN", "MN_EARN_WNE_P10"]].copy()
```

## Definations

> 1. UNITID: "Unit ID for institution"
2. INSTNM: "Institution Name"
3. CITY: "City"
4. STABBR: "State postcode"
5. ZIP: "Zipcode"
6. ADM_RATE: "Admission Rate"
7. UGDS: "Enrollment of undergraduate certificate/degree-seeking students"
8. TUITIONFEE_IN: "In-state tuition fee"
9. MN_EARN_WNE_P10: "Mean earnings of students working and not enrolled 10 years after entry"

## Created data frame "df3" with all 4 Maryland universities

```
uni_names = ['University of Maryland-Baltimore County', 'University of Maryland-College Park', 'Loyola University Maryland', 'Johns Hopkins University']
df3 = df2.loc[df2['INSTNM'].isin(uni_names)]
df3
```

```
# Minimum instate tuition fee of the 4 universities
df3["TUITIONFEE_IN"].min()
```

```
# Maximum in-state tuition fee of the 4 universities
df3["TUITIONFEE_IN"].max()
```

```
# Average in-state tuition fee of the 4 universities
df3["TUITIONFEE_IN"].mean()
```

```
# Created dataframe "df4 with 5 states universities
lst = ['MD', 'VA', 'DC', 'DE', 'PA']#State codes
df4 = df2.loc[df2['STABBR'].isin(lst)]# creating dataframe
df4.reset_index(inplace = True, drop = True) # resetting index
df4
```

```
# Converted df4 to .csv file with name 'Five_States_Colleges.csv'
df4.to_csv('Five_States_Colleges.csv', index=False, header=True)
```
