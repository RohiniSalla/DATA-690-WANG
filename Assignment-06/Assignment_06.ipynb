Write a function named cleanse_year() that takes a string as input and removes any asterisks in the string and return the cleansed string.

```python
def cleanse_year(year):
    return year.replace("*","")
```

Test the function by using test strings (for example, "1989*", "20*10")

```python
cleanse_year("1989*")
```

```python
cleanse_year("20*10")
```

Cleanse the column "Census Year" using the function cleanse_year() and Pandas's apply() function.

```python
df1['Census Year'] = df1['Census Year'].apply(cleanse_year)
df1
```

Repeat this process for column "Total Population", "Census Cost", "Average Cost Per Person". (define, test, and apply the function)

```python
import math

def cleanse_cost(raw_cost):
     # if value conatins '$'':
    cost = raw_cost.lstrip("$").replace(",","")
    
    # if value conatins Billion:
    if  cost.find("Billion") != -1:          
        cost = cost.replace('Billion','')
        cost = float(cost)* math.pow(10,8)
    return str(cost)
```

```python
cleanse_cost("$2,492,830,000") #testing
```

```python
cleanse_cost("$13 Billion")  #testing
```

```python
df1["Census Cost"] = df1["Census Cost"].apply(cleanse_cost)
df1
```

```python
def cleanse_population(popultaion):
  return int(popultaion.replace(',',''))
```

```python
cleanse_population("45,987")    #testing
```

```python
df1['Total Population'] = df1['Total Population'].apply(cleanse_population)
df1
```

```python
def cleanse_averagecost(cost):
  cost = cost.replace('$','')
  cost = cost.replace('\t','')
  cost = cost.replace('\n','')
  if 'cents' in cost:
    cost = cost.replace('cents','')
    cost = float(cost)/100
  return cost
```

```python
cleanse_averagecost("18.33 cents")
```

```python
cleanse_averagecost("$10.02")
```

```python
df1['Average Cost Per Person'] = df1['Average Cost Per Person'].apply(cleanse_averagecost)
df1
```
